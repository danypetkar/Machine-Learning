```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_style('whitegrid')
from sklearn.datasets import make_blobs
```

```python
centers = [[0.5,2],[-1,-1],[1.5,-1]]
```

```python
X ,_ = make_blobs(n_samples=400,centers=centers,cluster_std=0.5)
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1])
plt.show()
```

# DBSCAN

```python
# Density Based Spatial Clustering of Application with Noise

from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.4,min_samples=13)
```

```python
db.fit(X)
```

```python
identified_cluster = db.fit_predict(X)
identified_cluster
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=identified_cluster,cmap='rainbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)
kmeans.fit(X)
```

```python
identified_cluster1 = kmeans.fit_predict(X)
identified_cluster1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=identified_cluster1,cmap='rainbow')
plt.show()
```
