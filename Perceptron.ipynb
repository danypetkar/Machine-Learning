```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
X  = np.array([-1,0,1,2,3,4])
y = np.array([-3,-1,1,3,5,7])
```

```python
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(X.reshape(-1,1),y)
```

```python
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X.reshape(-1,1)),color='orange')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
lin_reg.predict(np.array([10]).reshape(-1,1))
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
import keras
```

```python
model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
model.fit(X,y,epochs=500)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X.reshape(-1,1)),color='orange')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
model.predict(np.array([10]).reshape(-1,1))
```
